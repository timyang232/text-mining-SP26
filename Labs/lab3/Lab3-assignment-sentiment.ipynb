### Sentence 1: "I love apples" — compound: 0.6369
The result is correct. "Love" has a high positive valence in the VADER lexicon, producing a positive compound score. No special rules are triggered and the output accurately reflects the sentiment.

### Sentence 2: "I don't love apples" — compound: -0.5216
The result is correct. VADER applies its negation rule, detecting "don't" and reversing the valence of "love". This correctly flips the sentiment to negative, showing that VADER handles simple negation well.

### Sentence 3: "I love apples :-)" — compound: 0.7579
The result is correct and stronger than Sentence 1. VADER recognizes ":-)" as a positive emoticon from its built-in emoticon lexicon and adds its valence on top of "love", boosting the compound score above that of Sentence 1.

### Sentence 4: "These houses are ruins" — compound: -0.4404
The result is incorrect. VADER assigns negative valence to "ruins" because the word is associated with destruction in its lexicon. However, here "ruins" is a noun referring to ancient remains, which is neutral. VADER cannot distinguish word senses or part-of-speech, leading to a wrong negative classification.

### Sentence 5: "These houses are certainly not considered ruins" — compound: 0.5867
The result is not necessarily right. The negation of "ruins" flips its valence to positive, and "certainly" acts as an intensifier that amplifies the score. However, the sentence is a neutral descriptive statement. The positive output is an artifact of rule interaction rather than genuine sentiment understanding.

### Sentence 6: "He lies in the chair in the garden" — compound: -0.4215
The result is incorrect. VADER assigns negative valence to "lies" because the word is associated with dishonesty in its lexicon. Here, however, "lies" means "to lie down", which is entirely neutral. This is a clear case of lexical ambiguity that VADER cannot resolve, as it has no mechanism to consider context or word sense.

### Sentence 7: "This house is like any house" — compound: 0.3612
The result is incorrect. VADER treats "like" as a positive sentiment word (as in "to like something") and assigns it positive valence. In this sentence "like" is used as a preposition meaning "similar to", carrying no sentiment. This again illustrates VADER's inability to handle lexical ambiguity.

````python
def vader_output_to_label(vader_output):
    """
    map vader output e.g.,
    {'neg': 0.0, 'neu': 0.0, 'pos': 1.0, 'compound': 0.4215}
    to one of the following values:
    a) positive float -> 'positive'
    b) 0.0 -> 'neutral'
    c) negative float -> 'negative'
    
    :param dict vader_output: output dict from vader
    
    :rtype: str
    :return: 'negative' | 'neutral' | 'positive'
    """
    compound = vader_output['compound']
    
    if compound < 0:
        return 'negative'
    elif compound == 0.0:
        return 'neutral'
    elif compound > 0.0:
        return 'positive'
    
assert vader_output_to_label( {'neg': 0.0, 'neu': 0.0, 'pos': 1.0, 'compound': 0.0}) == 'neutral'
assert vader_output_to_label( {'neg': 0.0, 'neu': 0.0, 'pos': 1.0, 'compound': 0.01}) == 'positive'
assert vader_output_to_label( {'neg': 0.0, 'neu': 0.0, 'pos': 1.0, 'compound': -0.01}) == 'negative'
````

````python
# Question 3a
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from sklearn.metrics import classification_report

# Load tweets
with open('my_tweets.json') as f:
    my_tweets = json.load(f)

# Load VADER model
vader_model = SentimentIntensityAnalyzer()

tweets = []
all_vader_output = []
all_vader_labels = []
gold = []

# settings (to change for different experiments)
to_lemmatize = True 
pos = set()

for id_, tweet_info in my_tweets.items():
    the_tweet = tweet_info['text_of_tweet']
    vader_output = vader_model.polarity_scores(the_tweet) # run vader
    vader_label = vader_output_to_label(vader_output) # convert vader output to category
    
    tweets.append(the_tweet)
    all_vader_output.append(vader_output)
    all_vader_labels.append(vader_label)
    gold.append(tweet_info['sentiment_label'])
    
# use scikit-learn's classification report
print(classification_report(gold, all_vader_labels, digits=3))
````

### Question 3a quantitative evaluation

The classification report provides Precision, Recall, and F1-score per class, plus macro and weighted averages.

**Results interpretation:**

- Positive (F1 = 0.512): VADER performs best here. Positive words like excellent, awesome, love are strongly represented in the VADER lexicon and easy to detect. However, the recall score (0.647) significantly exceeds the precision score (0.423), meaning VADER over predicts positive, it flags many neutral and negative tweets as positive.
- Negative (F1 = 0.400): Moderate performance. VADER catches some negative tweets but misses many where the negativity is expressed through context like frustration and sarcasm rather than explicit negative words.
- Neutral (F1 = 0.273): Worst performance. The recall score is very low (0.188), indicating VADER rarely predicts "neutral", it tends to push tweets into positive or negative based on any sentiment word found in the text, even when the tweet as a whole is neutral.
- Macro average (F1 = 0.395): This averages the three class F1 scores equally, regardless of class size. Because our dataset is balanced (17/17/16), this is the most honest single summary of VADER's overall performance. Crucially, the neutral class's poor F1 (0.273) pulls the macro average down just as much as the positive class's relatively better F1 (0.512) pulls it up, nothing is hidden. The result, 0.395, is closer to the worst-performing class than the best, which tells you VADER's performance is uneven rather than consistently moderate.
- Weighted average (F1 = 0.397): This weights each class's F1 by how many examples it has before averaging. Because the dataset is nearly balanced, the weighted average comes out almost identical to the macro average (0.397 vs 0.395).  

The overall accuracy of 42%, which is only marginally above the random chance baseline of 33% for a 3-class problem, indicating VADER provides limited value on this dataset, particularly due to its poor performance on classifying neutral.

**Most Relevant Scores**

The macro F1 (0.395) and per-class F1 are most informative, since the dataset is roughly balanced across three classes. The macro F1 avoids hiding poor neutral performance behind strong positive performance. Accuracy alone is less useful because it equally weights a correct neutral prediction and a correct positive prediction.

Train the scikit-learn classifier (Naive Bayes) using the airline tweets.

````python
import pathlib
import zipfile
from collections import Counter

import nltk
from nltk.corpus import stopwords
from sklearn.datasets import load_files
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

# Make sure the NLTK resources we need are available
for find_path, pkg in [
    ('tokenizers/punkt_tab', 'punkt_tab'),
    ('tokenizers/punkt', 'punkt'),
    ('corpora/stopwords', 'stopwords'),
]:
    try:
        nltk.data.find(find_path)
    except LookupError:
        nltk.download(pkg, quiet=True)

# Locate (and if needed, unzip) the airlinetweets dataset.
# The dataset ships as airlinetweets.zip in this folder; the unzipped layout is
#   airlinetweets/{negative,neutral,positive}/*.txt
candidate_paths = [
    pathlib.Path('airlinetweets'),
    pathlib.Path('airlinetweets') / 'airlinetweets',
]
airline_path = next((p for p in candidate_paths if (p / 'positive').exists()), None)
if airline_path is None and pathlib.Path('airlinetweets.zip').exists():
    with zipfile.ZipFile('airlinetweets.zip') as zf:
        zf.extractall('.')
    airline_path = next((p for p in candidate_paths if (p / 'positive').exists()), None)
assert airline_path is not None, 'Could not find the airlinetweets dataset.'
print('Dataset path:', airline_path)

airline_tweets = load_files(
    str(airline_path),
    categories=['negative', 'neutral', 'positive'],
    encoding='utf-8',
    decode_error='replace',
    allowed_extensions=['.txt'],
)
print('Total tweets:', len(airline_tweets.data))
print('Per-class counts:',
      {airline_tweets.target_names[i]: c
       for i, c in Counter(airline_tweets.target).items()})

# 80/20 stratified split, shared across all six experiments so the comparison
# is fair (only the vectoriser/min_df changes between runs).
X_train, X_test, y_train, y_test = train_test_split(
    airline_tweets.data,
    airline_tweets.target,
    test_size=0.20,
    random_state=42,
    stratify=airline_tweets.target,
)


def run_experiment(representation, min_df):
    """Train Multinomial NB with the given setting and print its classification report."""
    vec = CountVectorizer(
        min_df=min_df,
        tokenizer=nltk.word_tokenize,
        stop_words=stopwords.words('english'),
    )
    X_train_counts = vec.fit_transform(X_train)
    X_test_counts = vec.transform(X_test)

    if representation == 'tfidf':
        tfidf = TfidfTransformer()
        X_train_feats = tfidf.fit_transform(X_train_counts)
        X_test_feats = tfidf.transform(X_test_counts)
    else:  # 'count' = Bag of Words
        X_train_feats = X_train_counts
        X_test_feats = X_test_counts

    clf = MultinomialNB().fit(X_train_feats, y_train)
    y_pred = clf.predict(X_test_feats)

    header = f'{representation:>5} | min_df={min_df:<2} | vocab={len(vec.vocabulary_):>5}'
    print('\n=== ' + header + ' ===')
    print(classification_report(
        y_test, y_pred,
        target_names=airline_tweets.target_names,
        digits=3,
        zero_division=0,
    ))


# Six experiments: {airline_tfidf, airline_count} x {min_df=2, 5, 10}
for representation in ('tfidf', 'count'):
    for md in (2, 5, 10):
        run_experiment(representation, md)
````

Part of that is just that there are more negative tweets to learn from (around 1750 of them, versus ~1490 positive and ~1515 neutral), so Naive Bayes has a slightly higher prior on "negative" and tends to fall back on it for borderline tweets. But the bigger reason is that complaints have a really obvious vocabulary — words like *delay*, *cancelled*, *hours*, *worst*, *never*, plus the negation token *n't* — and those words barely show up in the other two classes. The model can lean on them.

The hardest class is **neutral**, every time (F1 stays between 0.74 and 0.78). That makes sense if you skim the actual tweets: neutrals are mostly factual questions or follow-ups like "@jetblue DM sent" or "what's the status of flight 1234?", and the words in them (*flight*, airline names, question marks) show up just as often in the other two classes. There's no clean signal for the model to grab onto, so it ends up confusing neutrals with both positives and negatives.

On TF-IDF vs. bag-of-words, BoW wins everywhere, and the biggest gain is on the neutral class (F1 0.74 → 0.78 at `min_df=2`). I think the reason is that TF-IDF down-weights words that appear in lots of tweets, but on this dataset those high-document-frequency words (*thanks*, *delay*, *united*) are exactly the ones that actually carry sentiment. Squashing them hurts more than it helps. It also fits with how Naive Bayes works under the hood — it expects raw counts, so giving it counts directly just feels like the more natural pairing.

The `min_df` threshold matters surprisingly little. Going from 2 → 10 cuts the vocabulary by more than 4× (2480 → 568 features), but accuracy only drops about 1.5 points (0.829 → 0.814 with BoW). Most of the predictive power lives in the high-frequency words, which all three thresholds keep. The drop isn't even across classes though: negative F1 barely moves (0.866 → 0.846), but positive recall takes the biggest hit (0.822 → 0.779) — that's because positives rely on a long tail of compliment words ("amazing", "awesome", "appreciated") that occur only a handful of times each and get filtered out as we raise `min_df`. Neutral F1 actually nudges up a little at `min_df=10`, probably because cutting low-frequency noise makes the model less sure about positive/negative predictions, so more borderline tweets fall into the neutral bucket by default.

So the best of the six setups is **bag-of-words with `min_df=2`** (macro-F1 0.825). The recurring weak spot in all of them is the neutral class, and I don't think any amount of `min_df` tuning is going to fix that — it would need features the bag-of-words model just doesn't have, like negation handling, POS information, or a proper Twitter-aware tokenizer.

### Question 6b — Discussion

Looking at the top-80 features the model leans on per class:

The features I *expected* to see were the obvious ones. For the negative class, `n't`, `cancelled`, `service`, `customer`, `bag`, `plane`, `time`, and `flight` all show up near the top, which is basically the vocabulary of an airline complaint — passengers venting about cancellations, lost bags, bad service, and so on. The question mark `?` ranking high on negative also makes sense, since a lot of complaints are phrased as exasperated questions ("why is my flight delayed *again*?"). For positive, the gratitude words come out exactly where I'd expect: `thanks`, `thank`, `great`, `love`, `!`, even `)` (the closing paren from smileys). And for neutral, I expected the dataset to lean on logistical tokens like `flight`, `flights`, `?`, and `http` — tweets asking for status updates or pointing at a DM thread — which is what the list shows.

What I *didn't* expect, or at least didn't want to see, is how much of every class list is taken up by stuff that has nothing to do with sentiment. The airline handles (`united`, `americanair`, `southwestair`, `jetblue`, `usairways`, `virginamerica`) are near the top of all three classes. `united` in particular is the #2 negative feature, and that's not because the word "united" means anything bad — it's just that United gets way more complaints in this dataset than the other airlines, so the model has effectively learned "if it mentions @united, predict negative." That's frequency bias, not sentiment. The same thing is happening with bare punctuation (`@`, `.`, `!`, `?`, `:`, `#`, `''`, `` `` ``, `-`): the NLTK tokenizer splits each of those out as its own token and the default English stop-word list doesn't filter them, so they look "important" just because they're everywhere. And tokens like `http`, `'s`, and `''` are showing up only because of how the tokenizer chops things up, not because they carry meaning.

So if I were trying to improve this model, I'd start by stripping or normalizing the noise. The handles are the single biggest issue — I'd either remove them entirely or replace each one with a generic `<AIRLINE>` placeholder so the model can't shortcut its way to "negative" by spotting "united". I'd also extend the stop-word list to drop the bare punctuation tokens (`.`, `,`, `:`, `''`, `` `` ``, `-`, `#`, `@`), strip URLs with a quick regex (`re.sub(r"https?://\S+|@\w+", "", text)`), and probably drop number/time/flight-number tokens too, since those mostly just push tweets toward neutral without adding any real signal.

The features I'd absolutely keep are the actual content words on each list — *delay*, *cancelled*, *worst*, *bad*, *rude*, *late*, *hours*, *never*, *service*, *customer*, *bag*, *plane* on the negative side, and *thanks*, *thank*, *great*, *love*, *awesome*, *best*, *help* on the positive side. I'd also keep `n't`: negation is one of the strongest sentiment cues a bag-of-words model can possibly use, and the tokenizer is already isolating it nicely. It would be even better to merge it back onto the verb during preprocessing (`did n't` → `didn't` / `did_not`) so the model sees something like "didn't_help" or "didn't_arrive" as a single token. The exclamation mark is the one borderline call — it shows up about 1040 times in positive tweets and only ~360 in negative, so it clearly carries signal, but I'd only keep it if I also added an emoticon-aware tokenizer so things like `:)` and `:-(` get treated the same way.

The big-picture takeaway is that a real chunk of the model's "top features" are dataset artifacts — airline names, punctuation, mentions, URLs — rather than sentiment markers. Cleaning those up, plus handling negation properly, should help the minority classes (positive and neutral) the most, because the model would stop using `united` as a near-synonym for "negative" and would have to actually look at what the tweet is saying.
