## [total points: 4] Exercise 1: NLTK
In this exercise, we use NLTK to apply **Part-of-speech (POS) tagging**, **Named Entity Recognition (NER)**, and **Constituency parsing**. The following code snippet already performs sentence splitting and tokenization. 

```python
import nltk
from nltk.tokenize import sent_tokenize
from nltk import word_tokenize
nltk.download('all')
```

```python
sentences_nltk = sent_tokenize(text)
```

```python
tokens_per_sentence = []
for sentence_nltk in sentences_nltk:
    sent_tokens = word_tokenize(sentence_nltk)
    tokens_per_sentence.append(sent_tokens)
```

We will use lists to keep track of the output of the NLP tasks. We can hence inspect the output for each task using the index of the sentence.

```python
sent_id = 1
print('SENTENCE', sentences_nltk[sent_id])
print('TOKENS', tokens_per_sentence[sent_id])
```

### [point: 1] Exercise 1a: Part-of-speech (POS) tagging
Use `nltk.pos_tag` to perform part-of-speech tagging on each sentence.

Use `print` to **show** the output in the notebook (and hence also in the exported PDF!).

```python
pos_tags_per_sentence = []
for tokens in tokens_per_sentence:
    curr_postag = nltk.pos_tag(tokens)
    pos_tags_per_sentence.append(curr_postag)
    print(curr_postag)
```

```python
print(pos_tags_per_sentence)
```

### [point: 1] Exercise 1b: Named Entity Recognition (NER)
Use `nltk.chunk.ne_chunk` to perform Named Entity Recognition (NER) on each sentence.

Use `print` to **show** the output in the notebook (and hence also in the exported PDF!).

```python
from nltk.chunk import ne_chunk
ner_tags_per_sentence = []

for tokens_pos_tagged in pos_tags_per_sentence:
    tokens_pos_tagged_and_named_entities = ne_chunk(tokens_pos_tagged)
    ner_tags_per_sentence.append(tokens_pos_tagged_and_named_entities)
    print("")
    print('ORIGINAL SENTENCE', tokens_pos_tagged)
    print('NAMED ENTITY RECOGNITION OUTPUT', tokens_pos_tagged_and_named_entities)
```

```python
print(ner_tags_per_sentence)
```

### [points: 2] Exercise 1c: Constituency parsing
Use the `nltk.RegexpParser` to perform constituency parsing on each sentence.

Use `print` to **show** the output in the notebook (and hence also in the exported PDF!).

```python
constituent_parser = nltk.RegexpParser('''
NP: {<DT>? <JJ>* <NN>*} # NP
P: {<IN>}           # Preposition
V: {<V.*>}          # Verb
PP: {<P> <NP>}      # PP -> P NP
VP: {<V> <NP|PP>*}  # VP -> V (NP|PP)*''')
```

```python
constituency_output_per_sentence = []

for tokens_pos_tagged in pos_tags_per_sentence:
    constituent_structure = constituent_parser.parse(tokens_pos_tagged)
    constituency_output_per_sentence.append(constituent_structure)

for tree in constituency_output_per_sentence:
    print(tree) # print the sentence structure
```

```python
print(constituency_output_per_sentence)
```

Augment the RegexpParser so that it also detects Named Entity Phrases (NEP), e.g., that it detects *Galaxy S III* and *Ice Cream Sandwich*

```python
constituent_parser_v2 = nltk.RegexpParser('''
NEP: {<NNP>+}       # Named Entity Phrase -> one or more proper nouns
NP: {<DT>? <JJ>* <NN>*} # NP
P: {<IN>}           # Preposition
V: {<V.*>}          # Verb
PP: {<P> <NP>}      # PP -> P NP
VP: {<V> <NP|PP>*}  # VP -> V (NP|PP)*''')
```

```python
constituency_v2_output_per_sentence = []
for tokens_pos_tagged in pos_tags_per_sentence:
    constituent_structure = constituent_parser_v2.parse(tokens_pos_tagged)
    constituency_v2_output_per_sentence.append(constituent_structure)

for tree in constituency_v2_output_per_sentence:
    print(tree) # print the sentence structure
```

```python
print(constituency_v2_output_per_sentence)
```

### [points: 3] Exercise 3a: Part of speech tagging
Compare the output from NLTK and spaCy regarding part of speech tagging.

* To compare, you probably would like to compare sentence per sentence. Describe if the sentence splitting is different for NLTK than for spaCy. If not, where do they differ?
* After checking the sentence splitting, select a sentence for which you expect interesting results and perhaps differences. Motivate your choice.

```python
# Sentence splitting comparison
print(f"Number of NLTK sentences: {len(sentences_nltk)}")
print(f"Number of spaCy sentences: {len(sents)}")

print("\nNLTK sentences:")
for i, s in enumerate(sentences_nltk):
    print(f"  [{i}]: {s}")

print("\nspaCy sentences:")
for i, s in enumerate(sents):
    clean = s.text.strip()
    print(f"  [{i}]: {clean}")

# POS comparison for "The six phones and tablets affected..." sentence
target = "The six phones"
nltk_idx = next(i for i, s in enumerate(sentences_nltk) if target in s)
spacy_idx = next(i for i, s in enumerate(sents) if target in s.text)

print(f"\n=== POS Tag Comparison for sentence (NLTK idx={nltk_idx}, spaCy idx={spacy_idx}) ===\n")

spacy_tokens = [(t.text, t.tag_) for t in sents[spacy_idx] if t.text.strip()]
nltk_tokens = pos_tags_per_sentence[nltk_idx]

print(f"{'Token':<15} {'NLTK':<10} {'spaCy':<10} {'Match'}")
print("-" * 45)
differences = 0
for i in range(min(len(nltk_tokens), len(spacy_tokens))):
    nt, ntag = nltk_tokens[i]
    st, stag = spacy_tokens[i]
    match = "Yes" if ntag == stag else "NO  <--"
    if ntag != stag:
        differences += 1
    print(f"{nt:<15} {ntag:<10} {stag:<10} {match}")

print(f"\nTotal differences: {differences}")
```

**Sentence splitting:**

Both NLTK and spaCy produce exactly **6 sentences**, and the sentence boundaries are nearly identical. The only difference is in **quote handling** around sentence 2/3. Sentence 2 ends with `Apple."`, the period is inside the closing quotation mark. NLTK recognizes `."` as a sentence ending and attaches the closing quote to sentence 2, so sentence 3 cleanly starts with `In August...`. spaCy, however, splits the period and quotation mark into separate tokens and does not attach the closing `"` to sentence 2. Instead, that orphaned `"` gets pushed to the start of sentence 3, resulting in `" In August, Samsung lost...`. This shows that spaCy's sentence splitter struggles with punctuation sequences like `."` at sentence boundaries.

Both tools merge the URL with the "Documents filed..." paragraph into a single sentence (sentence 0). This is because sentence tokenizers rely on punctuation (`.`, `?`, `!`) to detect boundaries: the URL ending in `.html` is not recognized as a sentence-ending period, and the blank line (newline characters) between the URL and the paragraph is not treated as a sentence boundary either. This is a limitation when processing text that contains non-natural-language elements like URLs.

**Sentence choice:**

We selected sentence index 1 ("The six phones and tablets affected are the Galaxy S III...") because it contains many product names (Galaxy S III, Jelly Bean, Galaxy Tab), version numbers (8.9, 10.1), and a mix of proper nouns, common nouns, and adjectives, making it a good candidate for revealing POS tagging differences.

**POS tag comparison:**

For the selected sentence, the fine-grained POS tags (`tag_`) from both NLTK and spaCy are **identical** for all 41 tokens (0 differences). Both tools use the Penn Treebank tagset and fully agree on this sentence.

However, differences do appear in other sentences. For sentence 0 ("Documents filed to the San Jose federal court..."), where the URL causes different tokenization, comparing the shared tokens reveals:
- **"Jelly"**: NLTK → RB (adverb), spaCy → NNP (proper noun). spaCy is correct here since "Jelly Bean" is a proper noun (Android OS version name).
- **"Documents"**: NLTK → NNS (common plural noun), spaCy → NNPS (proper plural noun).
- **"filed"**: NLTK → VBN (past participle), spaCy → VBD (simple past). NLTK arguably captures the reduced relative clause ("Documents [that were] filed") better.
- **"operating"**: NLTK → VBG (gerund), spaCy → NN (noun). In "operating systems," spaCy treats it as part of a compound noun.

These differences show that while both tools use the same tagset, their underlying models can disagree on ambiguous words.

### [points: 2] Exercise 3b: Named Entity Recognition (NER)
* Describe differences between the output from NLTK and spaCy for Named Entity Recognition. Which one do you think performs better?

**Differences between NLTK and spaCy NER:**

1. **Entity type richness**: spaCy uses a much richer set of entity labels: ORG, GPE, DATE, MONEY, CARDINAL, NORP (nationality/religious/political group), and WORK_OF_ART. NLTK uses a smaller set: ORGANIZATION, PERSON, GPE, and GSP.

2. **Multi-word entities**: spaCy correctly groups multi-word entities (e.g., "November 23" as DATE, "Galaxy Tab 8.9 Wifi" as one entity, "South Korean" as NORP). NLTK typically recognizes only individual words or short spans (e.g., "Samsung" alone, "Galaxy" alone).

3. **Numeric/value entities**: spaCy recognizes numbers ("six" as CARDINAL), dates ("August" as DATE), and monetary values ("$1.05bn" as MONEY, "£0.66bn" as MONEY). NLTK misses almost all of these.

4. **Misclassifications**:
   - NLTK labels "Apple" as PERSON in some sentences instead of ORGANIZATION.
   - NLTK labels "Bean" (from "Jelly Bean") as GPE (a location).
   - NLTK labels "August" as GPE.
   - spaCy mislabels product names like "the Galaxy S III" and "Galaxy Tab 8.9 Wifi" as PERSON instead of a product entity.
   - spaCy labels "Jelly Bean" as WORK_OF_ART in one instance and ORG in another: inconsistent.

**Which performs better?**

spaCy performs significantly better overall. It recognizes more entities, groups multi-word entities correctly, uses a richer label set, and makes fewer critical misclassifications. While it incorrectly labels some Samsung product names as PERSON, its overall accuracy and coverage far exceed NLTK's rule-based NER module.

### [points: 2] Exercise 3c: Constituency/dependency parsing
Choose one sentence from the text and run constituency parsing using NLTK and dependency parsing using spaCy.
* describe briefly the difference between constituency parsing and dependency parsing
* describe differences between the output from NLTK and spaCy.

```python
# Choose "Samsung, which is the world's top mobile phone maker, is appealing the ruling."
target = "world"
nltk_idx = next(i for i, s in enumerate(sentences_nltk) if target in s)
spacy_idx = next(i for i, s in enumerate(sents) if target in s.text)

print("Selected sentence:", sentences_nltk[nltk_idx].strip())

print("\n=== NLTK Constituency Parsing (v2, with NEP rule) ===\n")
print(constituency_v2_output_per_sentence[nltk_idx])

print("\n=== spaCy Dependency Parsing ===\n")
print(f"{'Token':<15} {'Dep Label':<12} {'Explanation':<30} {'Head'}")
print("-" * 70)
for token in sents[spacy_idx]:
    if token.text.strip():
        explanation = spacy.explain(token.dep_) or ""
        print(f"{token.text:<15} {token.dep_:<12} {explanation:<30} {token.head.text}")

print("\n=== spaCy Dependency Tree Visualization ===")
displacy.render(sents[spacy_idx], jupyter=True, style='dep')
```

**Constituency parsing vs. dependency parsing:**

- **Constituency parsing** decomposes a sentence into nested sub-phrases (constituents) in a tree structure. Each internal node represents a phrase type (NP, VP, PP), and the leaves are individual words. It answers: *"How do words group into phrases?"*

- **Dependency parsing** identifies directed grammatical relationships between individual words. Each word (dependent) is connected to exactly one other word (head) via a labeled relation (e.g., nsubj, dobj, amod). It answers: *"Which word depends on which other word?"*

**Output differences for the selected sentence:**

NLTK's `RegexpParser` (v2, with our added NEP rule `{<NNP>+}`) produces a rule-based constituency tree. It captures phrase structures that match the regex patterns (NP, VP, PP, NEP), and the NEP rule successfully groups "Samsung" as a named entity phrase. However, many tokens are still left unchunked directly under the root `S` node — function words like "which" and the possessive "'s" are not grouped into any phrase. The output remains incomplete and shallow.

spaCy's dependency parser produces a fully connected graph where *every* token has an assigned head and dependency label. Using `spacy.explain()` (as taught in Lab1.3), we can interpret each label. It correctly identifies:
- "Samsung" as the subject (nsubj) of "appealing"
- "which" as subject (nsubj) of "is" in the relative clause
- "maker" as the attribute (attr) of the copular "is"
- "ruling" as the direct object (dobj) of "appealing"
- Possessive structure: "'s" marks "world" as the possessor of "maker"

The `displacy.render()` visualization (also from Lab1.3) makes the dependency structure much easier to read than the raw text output, clearly showing the directed arcs between tokens.

The spaCy output is more informative and accurate because it uses a trained statistical model that handles complex constructions (relative clauses, possessives, appositives), while NLTK's rule-based approach is limited to what the handwritten regex rules can capture.
